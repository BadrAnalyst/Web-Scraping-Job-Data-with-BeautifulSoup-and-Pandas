### Importing Libraries & Methods

```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

### Inputting the URL

```python
url = 'https://wuzzuf.net/search/jobs/?q=illustrator&a=hpb'
```

### Create a Client-based Request to get the URL

```python
client = urlopen(url)
```

### Getting the HTML Code of the Full Page

```python
html = client.read()
```

```python
html
```

### Closing the Request

```python
client.close()
```

### Creating an HTML Parser using BeautifulSoup

```python
soup = bs(html, "html.parser")
```

```python
soup
```

### Creating a Container for the Needed Data

```python
containers = soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

### Bringing it all together

```python
f = open(r"C:\Users\Utilisateur\Desktop\Daten 2024-25\D.A Mustafa Othman\wuzzuf-illustator.csv","w")
header = "job_title, company_name, job_type\n"
f.write(header)
```

```python
for container in containers:
    jtitle = container.findAll("h2",{"class": "css-m604qf"})
    job_title = jtitle[0].text.strip()
    
    cname = container.findAll("a",{"class": "css-17s97q8"})
    company_name = cname[0].text.strip()
    
    jtype = container.findAll("a",{"class": "css-n2jc4m"})
    job_type = jtype[0].text.strip()
    
    #print(job_title)
    #print(company_name)
    #print(job_type)
    #print()
    
    #print(job_title + "," + company_name + "," + job_type)
    #print()
    
    f.write(job_title + "," + company_name + "," + job_type + "\n")
f.close()
    
```
